# Examples of graphs created using ggplot2

This notebook contains examples of ggplots created using the inbuilt ggplot datasets

```r
# Load packages
library("ggplot2")
library("dplyr")
```

# How does hwy vary with car type?

```r
ggplot(mpg) + 
geom_point(aes(class, cty), colour= "blue") +
geom_point(aes(class, hwy), colour= "green") +
xlab("Class") +
ylab("cty = blue /hwy = green") +
ggtitle("Cty and Hwy for different vehicle classes ") +
theme_bw(base_size=18)
```

```r
ggplot(mpg) +
  geom_jitter(aes(hwy, cty, color=class, size=0.5)) +
ggtitle("Cty versus hwy for a variety of vehicle classes") + 
theme_classic(base_size=18) # Add theme 
```

```r
class <- mpg %>% 
  group_by(class) %>% 
  summarise(n = n(), hwy = mean(hwy))

options(repr.plot.width=8, repr.plot.height=6)
ggplot(mpg) + 
  geom_jitter(data=mpg, aes(class, hwy),
              width=0.1, size=1.5) +
  geom_point(data=class, aes(class, hwy),
             color="red", size=4) +
  annotate('text', x=class$class, y=10, label=c("n=5", "n=47","n=41", "n=11", "n=33", "n=35", "n=62"))+
  ggtitle("How does hwy vary with car type?")
```

# How does hwy vary with displ?

```r
ggplot(mpg, 
       aes(displ, hwy)) + 
geom_point(aes(colour = drv, shape = drv), size=3) +  # Increase dot size
  scale_colour_discrete("Drive train") +
  scale_shape_discrete("Drive train") + # Merge 2 legends
labs(title = "How Hwy varies with displ", subtitle = "An improved graph") + # Add title
theme_minimal(base_size=18)  # Improve style
```

# How does diamond price vary with clarity?

```r
# How does the distribution of price vary with clarity?
ggplot(diamonds, aes(clarity, price)) + 
geom_violin() +
ggtitle("How does diamond price vary with clarity")
```

# How is diamond price affected by carat and clarity?

```r
options(repr.plot.width=12, repr.plot.height=10)

beaut_graphs <- ggplot(diamonds,
                   aes(x=carat,
                       y=price,
                       color=clarity)) + 
  theme_bw(base_size=18) +
  geom_point(alpha=0.2) +
  facet_grid(rows = vars(cut)) +
  labs(title = "Beautiful graph", subtitle = "How diamond price is affected by its clarity and carat")

beaut_graphs
```

```r
more_beautiful_graph <- ggplot(diamonds, 
                     aes(x=cut,
                         y=price, colour= cut, fill=cut)) +
  geom_violin() +
  facet_grid(~(as.factor(round(diamonds$carat))), scales='free') +
  theme_minimal(base_size=18) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  labs(title = "A more beautiful graph?", subtitle = "How diamond price is affected by its clarity and carat")

more_beautiful_graph

```
